Je vais essayer de regarder un petit peu au niveau de la puissance, c'est-à-dire se baser sur autre chose qu'une sphère mdr

Donc l'idéal ça serait d'avoir une courbe de puissance pour notre sattelite afin de rester fidèle à la réalité. Notre nouvelle fonction d'intensité reçue au sol sera maintenant également fonction de l'angle entre le satellite et le point au sol. Donc va falloir calculer tout ça

Bien sûr ! Une expression couramment utilisée pour décrire la fonction de rayonnement d'une antenne directive est celle du diagramme de rayonnement dans le domaine spatial. Pour une antenne directive typique montée sur un satellite, le diagramme de rayonnement peut être approximé par une fonction de type gaussienne ou une fonction sinus-cardinal (sinc).

Voici une expression générale pour le diagramme de rayonnement d'une antenne directive dans le domaine spatial, en coordonnées sphériques (azimut et élévation) :

$$ P(\theta, \phi) = A \cdot \left| f(\theta, \phi) \right|^2$$

Où :

- $$P(\theta, \phi) $$ est la puissance rayonnée dans la direction spécifiée par les angles d'azimut $$ \theta $$ et d'élévation $$ \phi $$.
- $$ A $$ est une constante de normalisation pour assurer que la puissance rayonnée totale est égale à la puissance d'entrée de l'antenne.
- $$ f(\theta, \phi) $$ est la fonction de rayonnement qui dépend de la conception spécifique de l'antenne.

Pour une antenne directive, la fonction $$ f(\theta, \phi) $$ peut être représentée par une fonction gaussienne, une fonction sinus-cardinal (sinc) ou d'autres fonctions spécifiques à la conception de l'antenne.

Par exemple, une expression simplifiée pour une antenne directive avec un lobe principal étroit peut être donnée par :

$$ f(\theta, \phi) = e^{-\left( \frac{\theta^2}{2\sigma_{\theta}^2} + \frac{\phi^2}{2\sigma_{\phi}^2} \right)} $$

Où $$ \sigma_{\theta} $$ et $$ \sigma_{\phi} $$ sont les largeurs du lobe principal dans les directions d'azimut et d'élévation, respectivement.

Il est à noter que ces expressions simplifiées décrivent généralement le comportement idéal d'une antenne directive dans des conditions idéales et que la modélisation précise du diagramme de rayonnement peut nécessiter des techniques de simulation électromagnétique plus avancées.

```python
import numpy as np
import matplotlib.pyplot as plt

# Définition de la fonction de rayonnement
def radiation_pattern(theta, phi, sigma_theta, sigma_phi):
    return np.exp(-(theta**2 / (2 * sigma_theta**2) + phi**2 / (2 * sigma_phi**2)))

# Paramètres de l'antenne
sigma_theta = 30  # Largeur du lobe principal en azimut (en degrés)
sigma_phi = 15    # Largeur du lobe principal en élévation (en degrés)

# Création des valeurs d'angles d'azimut et d'élévation
theta_values = np.linspace(-np.pi, np.pi, 500)  # Angles d'azimut de -pi à pi radians
phi_values = np.linspace(-np.pi/2, np.pi/2, 500)    # Angles d'élévation de -pi/2 à pi/2 radians

# Calcul du diagramme de rayonnement en fonction de theta à phi = 0
radiation_theta = radiation_pattern(theta_values, 0, np.radians(sigma_theta), np.radians(sigma_phi))

# Calcul du diagramme de rayonnement en fonction de phi à theta = 0
radiation_phi = radiation_pattern(0, phi_values, np.radians(sigma_theta), np.radians(sigma_phi))

# Tracé du diagramme de rayonnement en fonction de theta à phi = 0
plt.figure(figsize=(8, 6))
plt.polar(theta_values, radiation_theta, label='Phi = 0')
plt.title('Diagramme de rayonnement en fonction de theta à phi = 0')
plt.legend()
plt.show()

# Tracé du diagramme de rayonnement en fonction de phi à theta = 0
plt.figure(figsize=(8, 6))
plt.polar(phi_values, radiation_phi, label='Theta = 0')
plt.title('Diagramme de rayonnement en fonction de phi à theta = 0')
plt.legend()
plt.show()

```
