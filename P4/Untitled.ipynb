```python

import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import numpy as np

# Supposons que vous avez les données de densité de population sous forme de tableau numpy
# Ici, je vais générer des données aléatoires à des fins d'exemple
lons = np.random.uniform(low=-180, high=180, size=1000)
lats = np.random.uniform(low=-90, high=90, size=1000)
population_density = np.random.uniform(low=0, high=100, size=1000)

# Création de la carte
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())

# Ajout des frontières des pays et des côtes
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.LAND, color='lightgray')

# Affichage des densités de population en tant que points colorés
sc = ax.scatter(lons, lats, c=population_density, cmap='plasma',
                transform=ccrs.PlateCarree(), zorder=3, s = 10)
plt.colorbar(sc, label='Population Density')

# Ajouter des titres et des légendes
ax.set_title('Population Density Map')
plt.show()
```
